```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data.head(10)
```

```python
plt.scatter(data['year'],data['own'])
#adding title to the plot
plt.title("scatterplot")
#adding label to the plot
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

#line chart

```python
plt.plot(data['year'])
plt.plot(data['own'])
#adding title to the plot
plt.title("Line chart")
#adding label to the plot
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

#bar plot

```python
plt.bar(data['year'],data['own'])
#adding title to the plot
plt.title("Bar plot")
#adding label to the plot
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```
